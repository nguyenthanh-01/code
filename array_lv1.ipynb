```python
#1 Tìm phần tử đỉnh không nhỏ hơn phần tử lân cận của nó

def main():
    arr = [1, 3, 20, 4, 1, 0]

    for i in range(1, len(arr) - 1):
        if arr[i] < arr[i-1] or arr[i] < arr[i+1]:
            continue
        print(arr[i])
        return

main()
```

```python
#5 Phần tử nhỏ nhất/lớn thứ K trong mảng chưa sắp xếp

def main():
    arr = [12, 3, 5, 7, 19]
    k = 2

    print(sx_chon(arr, k))

def sx_chon(arr, k):
    for i in range(k):
        for j in range(i + 1, len(arr)):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
    return arr[k-1]

main()
```

```python
#6 Đếm số lần xuất hiện (hoặc tần suất) trong một mảng được sắp xếp

def main():
    arr = [1, 2, 2, 2, 2, 3, 4, 7 ,8 ,8]
    x = 2

    print(tk_np(arr, x))

def tk_np(arr, x):
    l = tk_np_(arr, x)
    r = tk_np_(arr, x, 1)
    return r - l + 1

def tk_np_(arr, x, tree = 0):
    l = 0
    r = len(arr) - 1

    if tree == 0:
        pos = -1
    else:
        pos = -2

    while l <= r:
        m = (r - l) // 2 + l

        if arr[m] == x:
            pos = m
            if tree == 0:
                r = m - 1
            else:
                l = m + 1

        elif arr[m] < x:
            l = m + 1
        elif arr[m] > x:
            r = m - 1

    return pos

main()
```

```python
#7 Sắp xếp một mảng 0, 1 và 2 | Vấn đề quốc kỳ Hà Lan

def main():
    arr = [0, 1, 1, 0, 1, 2, 1, 2, 0, 0, 0, 1]
    l = 0
    r = len(arr) - 1
    m = 0

    while m <= r:
        if arr[m] == 0:
            arr[m], arr[l] = arr[l], arr[m]
            l += 1
            m += 1
        elif arr[m] == 1:
            m += 1
        elif arr[m] == 2:
            arr[m], arr[r] = arr[r], arr[m]
            r -= 1
    print(*arr)

main()
```

```python
#8 Di chuyển tất cả các số âm về đầu và dương về cuối với khoảng trống thừa không đổi

def main():
    arr = [-12, 11, -13, -5, 6, -7, 5, -3, 11]
    l = 0
    r = len(arr) - 1

    while l <= r:
        if arr[l] < 0:
            l += 1
        elif arr[r] > 0:
            r -= 1
        else:
            arr[l], arr[r] = arr[r], arr[l]
    print(*arr)

main()
```

```python
#9 Sự kết hợp và giao nhau của hai mảng được sắp xếp

def main():
    arr1 = [1, 2, 4, 5, 6]
    arr2 = [2, 3, 5, 7]
    arr3 = []
    i = 0
    j = 0

    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            arr3 += [arr1[i]]
            i += 1
        else:
            arr3 += [arr2[j]]
            j += 1

    while i < len(arr1):
        arr3 += [arr1[i]]
        i += 1

    while j < len(arr2):
        arr3 += [arr2[j]]
        j += 1

    print(*arr3)

main()
```
